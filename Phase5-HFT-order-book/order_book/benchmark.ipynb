```python
import matplotlib.pyplot as plt

# Order volumes and execution times (in seconds)
order_sizes = [1000, 5000, 10000, 50000, 100000]
classic_times = [0.003, 0.020, 0.031, 0.189, 0.428]
optimized_times = [0.003, 0.017, 0.030, 0.197, 0.436]

# Create the plot
plt.plot(order_sizes, classic_times, marker='o', linestyle='-', label='Classic')
plt.plot(order_sizes, optimized_times, marker='o', linestyle='--', label='Optimized')

# Labeling
plt.xlabel("Number of Orders")
plt.ylabel("Execution Time (seconds)")
plt.title("HFT Order Book Performance: Classic vs Optimized")
plt.grid(True)
plt.legend()

# Display the chart
plt.show()
```
